```python
#libraries
import pandas as pd
```

```python
# load CSV file into DataFrame
def load_data(path):
    return pd.read_csv(path)

df = pd.read_csv("customer_analysis_clean.csv")
df
```

```python
# Drop duplicates and reset index, whereas all the values per caption have to be identical and the latter one is removed
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html
def dd(df):
    df = df.drop_duplicates().reset_index() # Dropping dupplicates and creating a new index column
    df.drop(columns=["index"], inplace=True) # Deleting the entire old index column
    return df
```

```python
# Execute DD Function in the Pipeline
df = dd(df)
```

```python
# Execute data formating Function in the Pipeline
def dateformating(df):
    df["effective_to_date"]=pd.to_datetime(df["effective_to_date"], errors="coerce") # change formate to date
    return df
```

```python
# Execute dataformating Function in the Pipeline
df = dateformating(df)
```

```python
df.sort_values(by=["customer_lifetime_value"])
```

```python
# do multiple data transformation and column changes/renaming

def transform_data(df):
    df["customer_lifetime_value"]=round(df["customer_lifetime_value"],0) # remove decimals
    df["total_claim_amount"]=round(df["total_claim_amount"],0) # remove decimals
    df.rename(columns={"month_effective_to_date":"month"}, inplace=True) # change long column name to shorter one 
    df["year"]=list(map(lambda date:date.strftime(format="%Y"),df["effective_to_date"])) # Create a new column with year derived from effective_to_date (%B would be for months)
    # https://www.tutorialspoint.com/c_standard_library/c_function_strftime.htm
    return df
```

```python
# Execute data transform Function in the Pipeline
df=transform_data(df)
```

```python
# Check for Null values in columns
df.isna().sum()
```

```python
# Other way to look for Null Values
df.info()
```

```python
# Drop Null Values
# 1) It has been isolated (via google sheets) that 613 nullvalues in "State" correspond exactly with nullvalues in "Response" column
# 1) plus 29 times with the null values in both vehicle_size and vehicle_class
# 2) There are remaining 579 nullvalues for both aforementioned columns vehicle_size and vehicle_class

def drop_null(df):
    df = df.dropna().reset_index() #drop the index (or raw) entirely if any cell in that index/raw is empty (by default like this)
    df.drop(columns=["index"], inplace=True) # Deleting the entire old index column
# https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.DataFrame.dropna.html
    return df
```

```python
# Execute drop null value Function in the Pipeline
df = drop_null(df)
```

```python
# cross-checking if null values are still left -> no
df.info()
```

```python
# Look at the different unique values in the respective column
df["vehicle_type"].value_counts()
```

```python
# Rename values in vehicle type column from M & A to "manual" & "automatic"
def full_name(x):
    if not x == x:
        return x
    else:
        vehicle_type = {'M': 'Manual', 
'A': 'Automatic'}
        return vehicle_type[x]
```

```python
# Execute full name writting Function in the Pipeline
df["vehicle_type"]=df["vehicle_type"].apply(full_name)
```
