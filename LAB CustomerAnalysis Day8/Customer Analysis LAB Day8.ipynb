```python
# define a new column called Region

#df["Region"] = [" "]*df.shape[0]
#df.head()

print(df["state"].value_counts())
```

```python
# Mapping the States into Regions defining the respective function

def region(x):
    
    if x !=x:
        return x

    else:
        dct_ = {"California": "West Region",
           "Washington": "East" ,
           "Oregon": "North West" ,
            "Arizona": "Central" ,
            "Nevada": "Central"}
    
        return dct_[x]
    
```

```python
# Applying the above defined function into a new Column called Region

df["region"]=df["state"].apply(region)
```
