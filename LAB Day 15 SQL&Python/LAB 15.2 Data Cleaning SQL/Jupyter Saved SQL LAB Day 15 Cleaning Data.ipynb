### Import right Libraries

```python
import pandas as pd
import getpass
import sqlalchemy as sa
```

### Connect to SQL DB with PW

```python
driver = 'mysql+pymysql'
user = 'root'
password = getpass.getpass(prompt='Your password:'"")
ip = '127.0.0.1'
```

### Giving all the info to connect to the database

```python
connection_string = f'{driver}://{user}:{password}@{ip}'
```

### Use Mapper to connect to the database. The db_connection is gatewway to the db.

```python
db_connection = sa.create_engine(connection_string)
```

### Check whether we have access to all the DB in your large environment

```python
insp = sa.inspect(db_connection)
db_list = insp.get_schema_names()
print(db_list)
```

### Check whether things also work on DB by DB level

```python
result = db_connection.execute("Select * from marketing_customer_analysis.customer_data")
result
```

```python
for row in result:
    print(row)
```
