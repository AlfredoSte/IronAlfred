## Importing and general cleaning

Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### We define a series of functions for our data cleaning pipeline:

```python
#Function for reading the data:
def read_customer_data():
    return pd.read_csv("Data_Marketing_Customer_Analysis_Round2.csv")
```

```python
#Function for standardizing column names:
def standardize(df):
    new_columns = df.columns.str.lower().str.replace(" ", "_")
    df.columns = new_columns
    return df
```

```python
#Function for dropping customer and unnamed:_0 columns
def drop_2cols(df):
    df.drop(columns=["unnamed:_0", "customer"], inplace=True)
    return df
```

```python
#Drop duplicates and reset index
def dd(df):
    df = df.drop_duplicates().reset_index()
    df.drop(columns=["index"], inplace=True)
    return df
```

```python
#Replacing null values
def fill_null(df):
    df["number_of_open_complaints"] = df["number_of_open_complaints"].fillna(value=0)
    mean_months = round(df["months_since_last_claim"].mean())
    df["months_since_last_claim"] = df["months_since_last_claim"].fillna(value=mean_months)
    mean_income = round(df["income"].mean())
    df["income"] = df["income"].fillna(value=mean_income)
    return df
```

## Running the pipe

```python
df = read_customer_data().pipe(standardize).pipe(drop_2cols).pipe(dd).pipe(fill_null)
df.info()
```

### Turning the absolute numbers into Rates


```python
# just for the info again

rsp_by_channel = df3.groupby(['sales_channel','response']).size().reset_index(name="total")
rsp_by_channel
```

```python
# splitting the data into the different possiblities to see for all the sales channel, how many "yes responses" were there?

sales_channel_yes_response = rsp_by_channel[rsp_by_channel['response']=='yes']
sales_channel_no_response = rsp_by_channel[rsp_by_channel['response']=='no']

sales_channel_yes_response
```

```python
# ensured that the index numbers adjusted 

sales_channel_yes_response.reset_index()
```

```python
# creating a new table with the summed numbers of the responses

total_rsp_by_channel = rsp_by_channel.groupby(["sales_channel",])["total"].sum().reset_index()
total_rsp_by_channel
```

```python
# create a new data frame whether the ratios are calculated
# starting with bringing the relevant counter for the ratio calculation -> total pos. responses per sales_channel
sales_channel_response_final = sales_channel_yes_response.groupby(["sales_channel",])["total"].sum().reset_index()

# addting the relevant denumenator for the ratio calculation into the new df -> total amount of responses per sales channel
sales_channel_response_final ["total_resp"] =  total_rsp_by_channel["total"] 

# calculating the ratio between total and total_resp in a new column called total response ratio
sales_channel_response_final ["ratio"] = sales_channel_response_final["total"].div(sales_channel_response_final["total_resp"], axis=0).multiply(100)
sales_channel_response_final
```

```python
# plotting charts showing the (pos) response ratios per sales channel

fig2, ax2 = plt.subplots()
ax2.set_title("Number of Responses")

sns.barplot(x=sales_channel_response_final["sales_channel"], y = sales_channel_response_final["ratio"])
```
